### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



Calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos

```python
ingreso_tienda1=tienda["Precio"].sum()
print(f"Ingreso total de la tienda 1:{ingreso_tienda1:.2f}")
```

```python
ingreso_tienda2=tienda2["Precio"].sum()
print(f"Ingreso total de la tienda 2:{ingreso_tienda2:.2f}")
```

```python
ingreso_tienda3=tienda3["Precio"].sum()
print(f"Ingreso total de la tienda 3:{ingreso_tienda3:.2f}")
```

```python
ingreso_tienda4=tienda4["Precio"].sum()
print(f"Ingreso total de la tienda 4:{ingreso_tienda4:.2f}")
```

# 2. Ventas por categoría

Calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda

```python
ventas_por_categoria_tienda1 = tienda.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("Ventas por categoría en Tienda 1:")
display(ventas_por_categoria_tienda1)
```

```python
ventas_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("Ventas por categoría en Tienda 2:")
display(ventas_por_categoria_tienda2)
```

```python
ventas_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("Ventas por categoría en Tienda 3:")
display(ventas_por_categoria_tienda3)
```

```python
ventas_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("Ventas por categoría en Tienda 4:")
display(ventas_por_categoria_tienda4)
```

# 4. Productos más y menos vendidos

Identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualizar los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

# **Visualización**

```python
import matplotlib.pyplot as plt
```

## Ingreso Total por Tienda

```python
ingresos_totales = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso Total': [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]
})

print("Ingresos Totales por Tienda:")
display(ingresos_totales)
```

```python
calificaciones_promedio = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [calificacion_promedio_tienda1, calificacion_promedio_tienda2, calificacion_promedio_tienda3, calificacion_promedio_tienda4]
})

print("Calificaciones Promedio por Tienda:")
display(calificaciones_promedio)
```

## Costo de Envío Promedio por Tienda

```python
plt.figure(figsize=(10, 6))
plt.bar(envios_promedio['Tienda'], envios_promedio['Costo de Envío Promedio'], color=['violet', 'blue', 'green', 'yellow'])
plt.title('Costo de Envío Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de Envío Promedio (COP)')
plt.show()
```

## Ventas por Categoría por Tienda

```python
ventas_por_categoria_total = pd.concat([
    ventas_por_categoria_tienda1.rename('Tienda 1'),
    ventas_por_categoria_tienda2.rename('Tienda 2'),
    ventas_por_categoria_tienda3.rename('Tienda 3'),
    ventas_por_categoria_tienda4.rename('Tienda 4')
], axis=1).fillna(0).astype(int)

ventas_por_categoria_total.index.name = 'Categoría del Producto'
ventas_por_categoria_total = ventas_por_categoria_total.stack().reset_index()
ventas_por_categoria_total.columns = ['Categoría del Producto', 'Tienda', 'Cantidad de Ventas']

print("Ventas por Categoría por Tienda:")
display(ventas_por_categoria_total.head())
```

```python
ventas_pivot = ventas_por_categoria_total.pivot_table(
    index='Categoría del Producto',
    columns='Tienda',
    values='Cantidad de Ventas'
)

fig, ax = plt.subplots(figsize=(14, 8))
ventas_pivot.plot(kind='bar', ax=ax, colormap='viridis')
plt.title('Ventas por Categoría por Tienda')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad de Ventas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
